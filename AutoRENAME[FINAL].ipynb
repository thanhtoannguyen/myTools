```python
import os
import glob

def rename_files(folder_path):
    # Lấy danh sách file .mp4 trong thư mục
    search_pattern = os.path.join(folder_path, '*.mp4')
    files = glob.glob(search_pattern)

    for i, file_path in enumerate(files, start=1):
        # Tạo tên mới bằng cách thêm số hàng loạt vào sau 'todn_'
        new_file_name = f"todn_p31_{i}.mp4"

        # Đường dẫn mới đầy đủ
        new_file_path = os.path.join(folder_path, new_file_name)

        # Đổi tên file
        os.rename(file_path, new_file_path)
        print(f"Renamed: {file_path} -> {new_file_path}")

# Đường dẫn thư mục chứa file.mp4
folder_path = '/content/drive/MyDrive/TODN_LAB/VIDEO THÀNH VIÊN/Part_31'

# Gọi hàm đổi tên file hàng loạt
rename_files(folder_path)
```
