```python
import io
import os
from PIL import Image
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Get the path to the watermark image
watermark_path = '/content/drive/MyDrive/TODN_LAB/Resources/watermark.png'

# Get the path to the directory containing the images to be watermarked
images_dir = '/content/drive/MyDrive/TODN_LAB/Pics/Part1'

# Check if the images directory exists
if not os.path.exists(images_dir):
    print(f"Images directory '{images_dir}' does not exist.")

# Open the watermark image
watermark = Image.open(watermark_path)

# Get a list of all the images in the directory
images = os.listdir(images_dir)

# Loop through the images and add the watermark
for image in images:
    # Check if the file is a JPG image
    if not image.lower().endswith('.jpg'):
        continue  # Skip to the next iteration if the file is not a JPG image

    try:
        # Open the image
        image_path = os.path.join(images_dir, image)
        im = Image.open(image_path)

        # Get the width and height of the image
        width, height = im.size

        # Calculate the position of the watermark
        x = (width - watermark.width) // 2
        y = (height - watermark.height) // 2

        # Paste the watermark onto the image
        im.paste(watermark, (x, y), watermark)

        # Save the watermarked image over the original image
        im.save(image_path)

        print(f"Watermark added to {image}")
    except (IOError, OSError):
        print(f"Unable to process {image}. Skipped.")
```
