```python
import os
from PIL import Image

# Đường dẫn tới thư mục chứa hình ảnh
folder_path = "/content/drive/MyDrive/TODN_LAB/PIC_UNCROP"

# Lấy danh sách tất cả các tệp tin hình ảnh trong thư mục
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Duyệt qua từng tệp tin hình ảnh
for file_name in image_files:
    # Đường dẫn tới tệp tin hình ảnh
    image_path = os.path.join(folder_path, file_name)
    
    # Mở hình ảnh sử dụng thư viện Pillow
    image = Image.open(image_path)
    
    # Lấy kích thước của hình ảnh
    width, height = image.size
    
    # Tính số lượng pixel cần cắt bỏ từ phía dưới
    pixels_to_cut = int(height * 0.1)
    
    # Cắt bỏ 10% pixel từ phía dưới của hình ảnh
    cropped_image = image.crop((0, 0, width, height - pixels_to_cut))
    
    # Hiển thị ảnh đã cắt ra
    cropped_image.show()
    
    # Lưu đè lên ảnh ban đầu
    cropped_image.save(image_path)
```
